```python
# converting the existing column to pandas date-time
matches["date"] = pd.to_datetime(matches["date"])
```

```python
# creating predictors for machine learning models
# converting the venue (Home or Away) into categories and assigning it numbers (like 0 or 1)
matches['venue_code'] = matches['venue'].astype("category").cat.codes
```

```python
matches
```

```python
# converting the teams (20 teams) into categories and assigning it values
matches['opp_code'] = matches['opponent'].astype("category").cat.codes
```

```python
matches
```

```python
# basically take the hour in the time field and remove the colon and the minutes using regular exp.
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
# assigning a number to the day
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
# assign codes to W or L/D but to make it more complex, you can assign it 0, 1, or 2 for W, L, and D
# this will turn true into 1 and false into 0
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
matches
```

```python
# since our data doesn't have a linear relationship, we cannot use linear regression models
# for example, higher opp code doesn't mean the opponent is more difficult
# but RandomForestClassifier can pick up those trends that are not linear
from sklearn.ensemble import RandomForestClassifier
```

```python
# a random forest is a series of decision trees, but each decision tree has a slightly different parameter
# n_estimator parameter -- higher the number, longer this function will take but also potentially more accurate your model will be
# min_samples_split -- this is basically number of samples we want to have before we want to split the node
# sample_split prevents the trees from becoming too detailed -- won't split further if the current group of data points has fewer than 10 matches
# experiment with it since lowering that will decrease the accuracy of the model
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
# splitting our training and test data sets
# need to make sure all the data in the test set come after the data in the training set since it is timeseries
# need to predict the future using data in the past basically
train = matches[matches["date"] < '2022-01-01']
test = matches[matches["date"] >= '2022-01-01']
```

```python
predictors = ['venue_code', 'opp_code', 'hour', 'day_code']
```

```python
# fit our randomforestclassifier model -- train with predictors above, and try to predict the target
rf.fit(train[predictors], train['target'])
```

```python
preds = rf.predict(test[predictors])
print(preds)
```

```python
# determine the accracy of the model
# metric that will say if predicted a W, what percentage of the time did the team actually win and vice versa and combine that
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test['target'], preds)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual=test['target'], predicted=preds))
```

```python
# when we predicted a loss or a draw, we were correct of the time
# when we predicted a win, we were wrong most of the times than we were right
pd.crosstab(index=combined['actual'], columns=combined['predicted'])
```

```python
# improving precision using rolling averages
grouped_matches = matches.groupby('team')
```

```python
group = grouped_matches.get_group('Manchester City')
```

```python
group
```

```python
# rolling averages: if we are at match week 4, how did Man City do in the previous match weeks?
# pass the information to make better predictions
# closed="left" is very important e.g. if we want to predict week 3, it will take the rolling average of week 1, 2, and 3 and use it to predict 3
# it will be using knowledge of the future to predict the future so closed="left" avoids that
def rolling_averages(group, cols, new_cols):
  group = group.sort_values("date")
  rolling_stats = group[cols].rolling(3, closed='left').mean()
  group[new_cols] = rolling_stats
  # dropna removes all missing values in case we are in week 2 and we want to compute the rolling average of the previous 3 weeks, we can't so it will drop the missing the values
  group = group.dropna(subset=new_cols)
  return group
```

```python
from types import new_class
# we want to rolling averages for the following cols
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
print(new_cols)
```

```python
rolling_averages(group, cols, new_cols)
```

```python
# Now we can pass into our machine learning algorithm these new parameters and it will improve our accuracy
# basically group the teams, apply the rolling_averages function
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
# took the original dataframe, grouped it by team (creates one dataframe per team), computed the rolling average for every match
matches_rolling
```

```python
# drop the index level
matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
# our index has repeated numbers so we will just add a separate incremental index
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```

```python
def make_predictions(data, predictors): # common code so we will write it in a function and re-run it
  train = data[data["date"] < '2022-01-01']
  test = data[data["date"] >= '2022-01-01']
  rf.fit(train[predictors], train["target"])
  preds = rf.predict(test[predictors])
  combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
  precision = precision_score(test["target"], preds)
  return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
print(combined)
print(precision)
```

```python
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)
combined
```

```python
# if you look at the data above, opponent is Wolves but the team name is Wolverhamton Wanderers.
# so we will need to normalize the data
class MissingDict(dict):
  __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"}
mapping = MissingDict(**map_values)
```

```python
combined["new_team"] = combined["team"].map(mapping)
```

```python
# merging the dataframe with itself -- look for the new_team field and merge it with the opponent field on the other side
# basically seeing if the prediction e.g. between Arsenal v Burnley was the same as Burney v Arsenal
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
merged
```
